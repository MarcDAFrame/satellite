```python
def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return array[idx]
```

```python
def color_changer(img, var):
    final = []
#     final = np.array()
    colors_list = [0]
    for i in range(1,var):
        colors_list.append(255/i)

    for x,row in enumerate(img):
        final.append([])
        for y,color_value in enumerate(row):
#                 test = np.abs(color_value-colors_list).argmin()
#                 print(test)
                final[x].append(find_nearest(colors_list, color_value))
            
    return final
# final = color_changer(img_out)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img_out = np.array(color_changer(gray, 3))
# print(img_out)
img_out_path = path + 'out.png'
cv2.imwrite(img_out_path, img_out)
Image(img_out_path)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img_out = np.array(color_changer(gray, 50))
# print(img_out)
img_out_path = path + 'out.png'
cv2.imwrite(img_out_path, img_out)
Image(img_out_path)

edge_out = cv2.Canny(cv2.imread(img_out_path), 500, 501)
edge_out_path = path + 'edge_out.png'
cv2.imwrite(edge_out_path, edge_out)
```
